```python
import pandas as pd
```

```python
# Importando a primeira base de dados no excel
celulares_subtraidos1 = pd.read_excel('celularessubtraidos_2024_1_6_(1).xlsb')
```

```python
# Quantidade de dados não nulos preenchidos nas linhas das colunas, e também o seu tipo
celulares_subtraidos1.info()
```

```python
celulares_subtraidos2 = pd.read_excel('celularessubtraidos_2024_7_9_(1).xlsb')
```

```python
celulares_subtraidos2.info()
```

```python
# Concatenar as duas bases de dados.
# Dessa forma, conseguimos trabalhar melhor as informações preenchidas nas duas bases
df = pd.concat([celulares_subtraidos1, celulares_subtraidos2])
```

```python
# Validando se os dados foram concatenados de forma correta
df.info()
```

```python
df.head(15)
```

```python
# Visualizar os valores distindo das coluna DESDOBRAMENTO para que possamos validar o tipo de informação que temos
# para podermos tratar os dados nulos
df['DESDOBRAMENTO'].unique()
```

```python
# A função fillna nos possibilita tratar os dados nulos em determinada coluna.
# Nesse caso, iremos preencher os nulos com 'Não Informado'.
df['DESDOBRAMENTO'] = df['DESDOBRAMENTO'].fillna('Não Informado')
```

```python
# Como podemos visualizar, não possuímos mais dados nulos na coluna DESDOBRAMENTO
df.info()
```
